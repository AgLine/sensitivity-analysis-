```python
from bs4 import BeautifulSoup as bs
from openpyxl import Workbook
import requests
```

```python
url_ = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=30688&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false"

html = bs(requests.get(url_).content,"html.parser", from_encoding="utf-8")
```

```python
html
```

```python
cnt = html.select("body > div > div > div.score_total > strong > em")[0].contents[0].replace(',','')
print(cnt)
```

```python
def get_data(url):
    resp = requests.get(url)
    html = bs(resp.content, 'html.parser')
    score_result = html.find('div', {'class': 'score_result'})
    lis = score_result.findAll('li')
    for li in lis:
        scores = li.find('em').getText()
        review_text = li.find('p').getText().replace('관람객',"").replace("감상평 보기", "").replace("스포일러가 포함된 감상평입니다.", "").strip()
        if(scores == '5'):
            print(review_text)
```

```python
for x in range(1, int(cnt)//10+1):
    url = url_ + '&page=' + str(x)
    get_data(url)
```
