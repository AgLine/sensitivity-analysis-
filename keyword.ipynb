```python
from konlpy.tag import Okt
from collections import Counter
import matplotlib.pyplot as plt
```

```python
f = open('Tomboy.txt','r',encoding='utf-8')
```

```python
review = f.read()
```

```python
okt = Okt()
n = okt.nouns(review)
```

```python
for i,v in enumerate(n):
    if len(v)<2:
        n.pop(i)
count = Counter(n)

noun_list = count.most_common(150)

for v in noun_list:
    print(v)
```

```python
text = review
wc = WordCloud(stopwords = stopword, width=800, height=800, background_color="white", random_state=0, font_path=r'c:\Windows\Fonts\malgun.ttf')
plt.imshow(wc.generate(text))
plt.axis("off")
plt.savefig('Tomboy.png', dpi=500)
plt.show()
```
